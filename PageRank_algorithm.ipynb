```python
import numpy as np

# الگوریتم مطابق جزوه
def pagerank(graph):
    
    num_pages = len(graph)

    Q = np.zeros((num_pages, num_pages))
    for i, page_url in enumerate(graph.keys()):
        for j, link_url in enumerate(graph.keys()):
            if len(graph[link_url]) != 0:
                if page_url in graph[link_url]:
                    Q[i, j] = 1 / len(graph[link_url])

    d = np.zeros((num_pages, 1))
    for i, page_url in enumerate(graph.keys()):
        if len(graph[page_url]) == 0:
            d[i, 0] = 1
    e = np.ones((num_pages, 1))
    P = Q + (1/num_pages) * e @ d.T

    alpha = 0.85
    A = alpha * P + (1-alpha) * (1/num_pages) * e @ e.T

    eigenvalues, eigenvectors = np.linalg.eig(A)

    rank = eigenvectors[:, 0]

    rank = rank / np.sum(np.abs(rank))
    return A, rank
```

```python
import pandas as pd

def read_dataset(file_path):
    df = pd.read_csv(file_path)
    # ساخت گراف وب
    web_graph = {}

    for i, row in df.iterrows():
        if row['page_url'] not in web_graph:
            web_graph[row['page_url']] = []
        if not pd.isna(row['link_url']):
            web_graph[row['page_url']].append(row['link_url'])

            if row['link_url'] not in web_graph:
                web_graph[row['link_url']] = []

    return web_graph
```

```python
file_path = "sklearn-anatomy.csv"
web_graph = read_dataset(file_path)
```

```python
A, rank = pagerank(web_graph)
```

```python
rank = np.real(rank)
rank.shape
```
